```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker
```

```python
plt.rcParams.update({'font.size': 14, 'axes.titlesize': 14,'legend.loc': 'upper right', 'legend.fontsize': 11})

fig, axs = plt.subplots(2, 2, figsize=(10, 7), dpi=800)
fig.tight_layout(h_pad=3.0, w_pad=3.0)
fig.patch.set_facecolor('white')
##############################################################################
########################## DATASET SIZE EXPERIMENT
#############################################################################

# SPATIAL3D DATASET
spatial_partition = [2, 4, 6, 8, 10]
spatial_1 = [0.31, 0.25, 0.24, 0.23, 0.25]
spatial_2 = [1.12, 0.86, 0.66, 0.53, 0.54]
spatial_3 = [2.63, 1.68, 1.25, 1.05, 0.75]
spatial_4 = [8.64, 6.13, 3.62, 1.96, 1.45]
spatial_5 = [21.62, 16.93, 8.77, 4.62, 2.83]

axs[0, 0].plot(spatial_partition, spatial_1, label="$|D|$=25000", color='#130f40', marker='D',
         linewidth=1, markersize=4, markerfacecolor='#130f40')
axs[0, 0].plot(spatial_partition, spatial_2, label="$|D|$=50000", color='#27ae60', marker='o',
         linewidth=1, markersize=4, markerfacecolor='#27ae60')
axs[0, 0].plot(spatial_partition, spatial_3, label="$|D|$=100000", color='#c0392b', marker='v',
         linewidth=1, markersize=4, markerfacecolor='#c0392b')
axs[0, 0].plot(spatial_partition, spatial_4, label="$|D|$=200000", color='#4834d4', marker='x',
         linewidth=1, markersize=4, markerfacecolor='#4834d4')
axs[0, 0].plot(spatial_partition, spatial_5, label="$|D|$=400000", color='#6ab04c', marker='p',
         linewidth=1, markersize=4, markerfacecolor='#6ab04c')
   
axs[0, 0].set_ylabel("Execution Time (s)")
axs[0, 0].set_xlabel(r'$r$', fontsize=16)
axs[0, 0].set_xscale('linear')
axs[0, 0].set_yscale('log')
axs[0, 0].set_xticks(spatial_partition)
axs[0, 0].set_ylim(0.1, 1000)
axs[0, 0].yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
axs[0, 0].set_title("(a) SPATIAL3D")
axs[0, 0].legend()

# IONO3D 2D DATASET
iono2d_partition = [10, 20, 30, 40, 50]
iono2d_1 = [1.31, 0.9, 0.97, None, None]
iono2d_2 = [3.29, 1.86, 1.65, 1.95, 2.46]
iono2d_3 = [9.1, 4.06, 2.78, 2.68, 3.09]
iono2d_4 = [24.47, 10.53, 5.89, 4.7, 4.2]
iono2d_5 = [64.06, 26.92, 15.95, 10.63, 9.48]

axs[0, 1].plot(iono2d_partition, iono2d_1, label="$|D|$=50000", color='#130f40', marker='D',
         linewidth=1, markersize=4, markerfacecolor='#130f40')
axs[0, 1].plot(iono2d_partition, iono2d_2, label="$|D|$=100000", color='#27ae60', marker='o',
         linewidth=1, markersize=4, markerfacecolor='#27ae60')
axs[0, 1].plot(iono2d_partition, iono2d_3, label="$|D|$=200000", color='#c0392b', marker='v',
         linewidth=1, markersize=4, markerfacecolor='#c0392b')
axs[0, 1].plot(iono2d_partition, iono2d_4, label="$|D|$=400000", color='#4834d4', marker='x',
         linewidth=1, markersize=4, markerfacecolor='#4834d4')
axs[0, 1].plot(iono2d_partition, iono2d_5, label="$|D|$=800000", color='#6ab04c', marker='p',
         linewidth=1, markersize=4, markerfacecolor='#6ab04c')
   
axs[0, 1].set_ylabel("Execution Time (s)")
axs[0, 1].set_xlabel(r'$r$', fontsize=16)
axs[0, 1].set_xscale('linear')
axs[0, 1].set_yscale('log')
axs[0, 1].set_xticks(iono2d_partition)
axs[0, 1].set_ylim(0.1, 1000)
axs[0, 1].yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
axs[0, 1].set_title("(b) IONO3D")
axs[0, 1].legend()

##############################################################################
########################## EPS EXPERIMENT
#############################################################################

# SPATIAL3D DATASET
spatial_eps = [2, 4, 6, 8, 10]
spatial_1 = [57.8, 36.22, 19.6, 11.75, 7.94]
spatial_2 = [57.47, 41.22, 22.93, 12.04, 7.79]
spatial_3 = [38.87, 31.06, 16.64, 9.55, 5.77]
spatial_4 = [19.89, 19.25, 10.88, 5.04, 3.46]
spatial_5 = [15.29, 10.76, 7.33, 3.59, 2.55]

axs[1, 0].plot(spatial_eps, spatial_1, label="ε=0.02", color='#130f40', marker='D',
         linewidth=1, markersize=4, markerfacecolor='#130f40')
axs[1, 0].plot(spatial_eps, spatial_2, label="ε=0.04", color='#27ae60', marker='o',
         linewidth=1, markersize=4, markerfacecolor='#27ae60')
axs[1, 0].plot(spatial_eps, spatial_3, label="ε=0.06", color='#c0392b', marker='v',
         linewidth=1, markersize=4, markerfacecolor='#c0392b')
axs[1, 0].plot(spatial_eps, spatial_4, label="ε=0.08", color='#4834d4', marker='x',
         linewidth=1, markersize=4, markerfacecolor='#4834d4')
axs[1, 0].plot(spatial_eps, spatial_5, label="ε=0.10", color='#6ab04c', marker='p',
         linewidth=1, markersize=4, markerfacecolor='#6ab04c')
   
axs[1, 0].set_ylabel("Execution Time (s)")
axs[1, 0].set_xlabel(r'$r$', fontsize=16)
axs[1, 0].set_xscale('linear')
axs[1, 0].set_yscale('log')
axs[1, 0].set_xticks(spatial_eps)
axs[1, 0].set_ylim(1, 1000)
axs[1, 0].yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
axs[1, 0].set_title("(c) SPATIAL3D")
axs[1, 0].legend()



# IONO3D 2D DATASET
iono2d_eps = [10, 20, 30, 40, 50]
iono2d_1 = [37.65, 16.69, 10.32, 8.57, 7.57]
iono2d_2 = [24.27, 11.3, 7.17, 6.07, 5.84]
iono2d_3 = [16.85, 7.79, 4.98, 4.65, 4.73]
iono2d_4 = [13.34, 5.95, 4.22, 3.86, 3.96]
iono2d_5 = [8.9, 4.86, 3.36, 3.24, 3.69]

axs[1, 1].plot(iono2d_eps, iono2d_1, label="ε=0.50", color='#130f40', marker='D',
         linewidth=1, markersize=4, markerfacecolor='#130f40')
axs[1, 1].plot(iono2d_eps, iono2d_2, label="ε=0.75", color='#27ae60', marker='o',
         linewidth=1, markersize=4, markerfacecolor='#27ae60')
axs[1, 1].plot(iono2d_eps, iono2d_3, label="ε=1.00", color='#c0392b', marker='v',
         linewidth=1, markersize=4, markerfacecolor='#c0392b')
axs[1, 1].plot(iono2d_eps, iono2d_4, label="ε=1.25", color='#4834d4', marker='x',
         linewidth=1, markersize=4, markerfacecolor='#4834d4')
axs[1, 1].plot(iono2d_eps, iono2d_5, label="ε=1.50", color='#6ab04c', marker='p',
         linewidth=1, markersize=4, markerfacecolor='#6ab04c')
   
axs[1, 1].set_ylabel("Execution Time (s)")
axs[1, 1].set_xlabel(r'$r$', fontsize=16)
axs[1, 1].set_xscale('linear')
axs[1, 1].set_yscale('log')
axs[1, 1].set_xticks(iono2d_eps)
axs[1, 1].set_ylim(1, 1000)
axs[1, 1].yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
axs[1, 1].set_title("(d) IONO3D")
axs[1, 1].legend()


######################################################################################
figname = '3d_partition_selection_experiments'
plt.savefig(figname+'.png', bbox_inches='tight', dpi=600)
plt.show()
```
