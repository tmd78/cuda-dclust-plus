```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker
```

```python
plt.rcParams.update({'font.size': 14, 'axes.titlesize': 14,'legend.loc': 'upper center', 'legend.fontsize': 11})

fig, axs = plt.subplots(3, 2, figsize=(8, 8), dpi=800)
fig.tight_layout(h_pad=3.0, w_pad=3.0)
fig.patch.set_facecolor('white')
##############################################################################
########################## EPS EXPERIMENT
#############################################################################
# SPATIAL3D DATASET
spatial3d_eps = [0.02, 0.04, 0.06, 0.08, 0.1]
spatial3d_indexing_dbscan_runtime_eps = [7.1, 7.28, 5.62, 2.78, 1.78]
spatial3d_dclust_runtime_eps = [26.69, 28.63, 17.63, 12.86, 8.52]

axs[0, 0].plot(spatial3d_eps, spatial3d_indexing_dbscan_runtime_eps, label="CUDA-DClust+", color='#130f40', marker='D',
         linewidth=1, markersize=4, markerfacecolor='#130f40')
axs[0, 0].plot(spatial3d_eps, spatial3d_dclust_runtime_eps, label="CUDA-DClust", color='#c0392b', marker='v',
         linewidth=1, markersize=4, markerfacecolor='#c0392b')
   
axs[0, 0].set_ylabel("Execution Time (s)")
axs[0, 0].set_xlabel("ε")
axs[0, 0].set_xscale('linear')
axs[0, 0].set_yscale('log')
axs[0, 0].set_xticks(spatial3d_eps)
axs[0, 0].set_ylim(1, 100)
axs[0, 0].yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
axs[0, 0].set_title("(a) SPATIAL3D")

# IONO3D 2D
iono3d_eps = [0.5, 0.75, 1.0, 1.25, 1.5]
iono3d_indexing_dbscan_runtime_eps = [7.08, 5.39, 3.97, 3.32, 2.56]
iono3d_dclust_runtime_eps = [36.11, 29.8, 29.2, 27.67, 26.07]

axs[0, 1].plot(iono3d_eps, iono3d_indexing_dbscan_runtime_eps, label="CUDA-DClust+", color='#130f40', marker='D',
         linewidth=1, markersize=4, markerfacecolor='#130f40')
axs[0, 1].plot(iono3d_eps, iono3d_dclust_runtime_eps, label="CUDA-DClust", color='#c0392b', marker='v',
         linewidth=1, markersize=4, markerfacecolor='#c0392b')
   
axs[0, 1].set_ylabel("Execution Time (s)")
axs[0, 1].set_xlabel("ε")
axs[0, 1].set_xscale('linear')
axs[0, 1].set_yscale('log')
axs[0, 1].set_xticks(iono3d_eps)
axs[0, 1].set_ylim(1, 100)
axs[0, 1].yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
axs[0, 1].set_title("(b) IONO3D")

##############################################################################
########################## MINPTS EXPERIMENT
#############################################################################

# SPATIAL3D DATASET
spatial3d_minpts = [1, 2, 4, 8, 16]
spatial3d_indexing_dbscan_runtime_minpts = [2.93, 2.94, 2.68, 3.3, 4.91]
spatial3d_dclust_runtime_minpts = [13.5, 11.75, 12.3, 17.56, 26.7]

axs[1, 0].plot(spatial3d_minpts, spatial3d_indexing_dbscan_runtime_minpts, label="CUDA-DClust+", color='#130f40', marker='D',
         linewidth=1, markersize=4, markerfacecolor='#130f40')
axs[1, 0].plot(spatial3d_minpts, spatial3d_dclust_runtime_minpts, label="CUDA-DClust", color='#c0392b', marker='v',
         linewidth=1, markersize=4, markerfacecolor='#c0392b')

   
axs[1, 0].set_ylabel("Execution Time (s)")
axs[1, 0].set_xlabel("minPts")
axs[1, 0].set_xscale('log')
axs[1, 0].set_yscale('log')
axs[1, 0].set_xticks(spatial3d_minpts)
axs[1, 0].get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
axs[1, 0].set_ylim(1, 100)
axs[1, 0].yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
axs[1, 0].set_title("(c) SPATIAL3D")

# IONO3D 2D
iono3d_minpts = [4, 8, 16, 32, 64]
iono3d_indexing_dbscan_runtime_minpts = [2.76, 2.64, 2.6, 2.66, 2.67]
iono3d_dclust_runtime_minpts = [26.07, 26.07, 10.6, 14.03, 29.68]

axs[1, 1].plot(iono3d_minpts, iono3d_indexing_dbscan_runtime_minpts, label="CUDA-DClust+", color='#130f40', marker='D',
         linewidth=1, markersize=4, markerfacecolor='#130f40')
axs[1, 1].plot(iono3d_minpts, iono3d_dclust_runtime_minpts, label="CUDA-DClust", color='#c0392b', marker='v',
         linewidth=1, markersize=4, markerfacecolor='#c0392b')
   
axs[1, 1].set_ylabel("Execution Time (s)")
axs[1, 1].set_xlabel("minPts")
axs[1, 1].set_xscale('log')
axs[1, 1].set_yscale('log')
axs[1, 1].set_xticks(iono3d_minpts)
axs[1, 1].get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
axs[1, 1].set_ylim(1, 100)
axs[1, 1].yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
axs[1, 1].set_title("(d) IONO3D")


##############################################################################
########################## POINTS EXPERIMENT
#############################################################################

# SPATIAL3D DATASET
spatial3d_points = [25000, 50000, 100000, 200000, 400000]
spatial3d_indexing_dbscan_runtime_points = [0.25, 0.47, 0.83, 1.31, 2.97]
spatial3d_dclust_runtime_points = [1.32, 3.51, 5.28, 6.23, 12.09]

axs[2, 0].plot(spatial3d_points, spatial3d_indexing_dbscan_runtime_points, label="CUDA-DClust+", color='#130f40', marker='D',
         linewidth=1, markersize=4, markerfacecolor='#130f40')
axs[2, 0].plot(spatial3d_points, spatial3d_dclust_runtime_points, label="CUDA-DClust", color='#c0392b', marker='v',
         linewidth=1, markersize=4, markerfacecolor='#c0392b')
   
axs[2, 0].set_ylabel("Execution Time (s)")
axs[2, 0].set_xlabel(r'$|D|$')
axs[2, 0].set_xscale('log')
axs[2, 0].set_yscale('log')
axs[2, 0].set_xticks(spatial3d_points)
axs[2, 0].get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
axs[2, 0].set_ylim(0.01, 100)
axs[2, 0].yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
axs[2, 0].set_title("(e) SPATIAL3D")
axs[2, 0].ticklabel_format(style='sci', axis='x', scilimits=(0,0))

# IONO3D
iono3d_points = [100000, 200000, 400000, 800000, 1600000]
iono3d_indexing_dbscan_runtime_points = [0.98, 1.89, 2.61, 4.53, 10.41]
iono3d_dclust_runtime_points = [6.68, 9.57, 26.07, 44.32, 82.2]

axs[2, 1].plot(iono3d_points, iono3d_indexing_dbscan_runtime_points, label="CUDA-DClust+", color='#130f40', marker='D',
         linewidth=1, markersize=4, markerfacecolor='#130f40')
axs[2, 1].plot(iono3d_points, iono3d_dclust_runtime_points, label="CUDA-DClust", color='#c0392b', marker='v',
         linewidth=1, markersize=4, markerfacecolor='#c0392b')
   
axs[2, 1].set_ylabel("Execution Time (s)")
axs[2, 1].set_xlabel(r'$|D|$')
axs[2, 1].set_xscale('log')
axs[2, 1].set_yscale('log')
axs[2, 1].set_xticks(iono3d_points)
axs[2, 1].get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
axs[2, 1].set_ylim(0.1, 100)
axs[2, 1].yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
axs[2, 1].set_title("(f) IONO3D")
axs[2, 1].ticklabel_format(style='sci', axis='x', scilimits=(0,0))

######################################################################################
plt.subplots_adjust(top=0.92, bottom=0.08)

handles, labels = axs[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, ncol=5, labelspacing=16)


figname = '3d_all_experiments'
plt.savefig(figname+'.png', bbox_inches='tight', dpi=600)
plt.show()
```
